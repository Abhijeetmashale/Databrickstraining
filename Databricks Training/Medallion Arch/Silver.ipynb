```python
%sql
-- Clean transactions: remove metadata columns, deduplicate, add data quality checks
create or replace table dev.abhijeet_sliver.customer as
with cleaned_transactions as (
  select
    customer_id,
    customer_name,
    customer_type,
    industry,
    status,
    tier,
    phone,
    website,
    billing_city,
    billing_state,
    billing_country,
    created_date
  from dev.abhijeet_bronze.customer
  where customer_name is not null
    and customer_id is not null
    and created_date > date_sub(current_date(), 730)
)
select distinct *
from cleaned_transactions
```

```python
%sql
create or replace table dev.abhijeet_sliver.opportunity as
with cleaned_transactions1 as (
  select
    opportunity_id,
    customer_id,
    primary_contact_id,
    opportunity_name,
    stage,
    probability,
    currency,
    estimated_amount,
    expected_close_date,
    created_date,
    owner,
    lead_source
  from dev.abhijeet_bronze.opportunities
  where stage <> 'Closed Lost'
    and customer_id is not null
)
select distinct *
from cleaned_transactions1
```

```python
%sql
SELECT
  c.customer_name,
  c.customer_id,
  c.customer_type,
  c.industry,
  c.status,
  o.stage,
  o.currency,
  o.estimated_amount,
  o.expected_close_date
FROM dev.abhijeet_sliver.customer c
INNER JOIN dev.abhijeet_sliver.opportunity o
  ON c.customer_id = o.customer_id
```
