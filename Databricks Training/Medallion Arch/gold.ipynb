```python
%sql
create or replace view dev.abhijeet_gold.Customer_Tran   as  
select
  c.customer_name,
  o.stage,
  sum(o.estimated_amount)
  FROM dev.abhijeet_sliver.customer c
INNER JOIN dev.abhijeet_sliver.opportunity o
  ON c.customer_id = o.customer_id group by c.customer_name, o.stage order by sum(o.estimated_amount) desc
```

```python
%sql
create or replace view dev.abhijeet_gold.Customer_data   as  
select
  c.customer_name,
  c.customer_id,
  c.customer_type,
  c.industry,
  c.status,
  o.stage,
  o.currency,
  o.estimated_amount,
  o.expected_close_date
FROM dev.abhijeet_sliver.customer c
INNER JOIN dev.abhijeet_sliver.opportunity o
  ON c.customer_id = o.customer_id 
```

```python
%sql
select * from dev.abhijeet_gold.customer_data;
select * from dev.abhijeet_gold.Customer_Tran;
```
