```sql
SELECT
  c.customerID,
  c.first_name,
  c.last_name,
  c.email_address,
  c.city,
  c.state,
  c.country,
  t.transactionID,
  t.dateTime,
  t.product,
  t.quantity,
  t.unitPrice,
  t.totalPrice,
  t.paymentMethod
FROM samples.bakehouse.sales_customers c
JOIN samples.bakehouse.sales_transactions t
  ON c.customerID = t.customerID
```
